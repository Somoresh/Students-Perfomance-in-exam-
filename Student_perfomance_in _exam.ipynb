```python
# significant difference between male and female students performance in math, reading, and writing scores.




import scipy.stats as stats
data=pd.DataFrame(df)

male_scores = data[data['gender'] == 'male']
female_scores = data[data['gender'] == 'female']

alpha = 0.05 

for subject in ['math score', 'reading score', 'writing score']:
    male_subject_scores = male_scores[subject]
    female_subject_scores = female_scores[subject]
    
    t_stat, p_value = stats.ttest_ind(male_subject_scores, female_subject_scores)
    
    if p_value < alpha:
        print(f"There is difference between male and female students scores in {subject}.")
    else:
        print(f"There is no significant difference between male and female students scores in {subject}.")
        
        
        
```
